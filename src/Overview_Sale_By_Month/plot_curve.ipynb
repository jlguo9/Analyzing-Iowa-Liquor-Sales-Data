# Plot the Curve of History Sale Data by Month

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import glob

cur_path = os.getcwd()

path = os.path.join(cur_path, "sales_by_month")
csv_files = glob.glob(os.path.join(path, "*.csv"))

data = pd.read_csv(csv_files[0])

data['date'] = pd.to_datetime(data[['year', 'month']].assign(day=1))

fig, ax = plt.subplots()
data.plot(x="date", y="sale", kind="line", ax=ax)
ax.set_xlabel('Date')
ax.set_title('History Sale Data by Month')
plt.savefig('./graphs/history_sale.png', bbox_inches='tight')
plt.show()
```

# Plot Predicted Sale Next Year (2023) by Month

```python
path = os.path.join(cur_path, "GBT_pred")
csv_files = glob.glob(os.path.join(path, "*.csv"))

pred = pd.read_csv(csv_files[0])

pred['date'] = pd.to_datetime(pred[['year', 'month']].assign(day=1))

fig, ax = plt.subplots()
pred.plot(x="date", y="prediction", kind="line", ax=ax)
ax.set_xlabel('Date')
ax.set_title('Predicted Sale Data in 2023 by Month')
plt.savefig('./graphs/prediction.png', bbox_inches='tight')
plt.show()

```
